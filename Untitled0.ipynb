```python
lowResImages = np.array(lowResImages)
```

```python
categories = np.array(cats)
```

```python
xtrain = lowResImages/255
ytrain = k.utils.to_categorical(categories)
```

```python
hiResTest=np.load("/content/gdrive/MyDrive/Data/hiResTest.npy")
catsTest = np.load("/content/gdrive/MyDrive/Data/categoriesTest.npy")
```

```python
lowResTest=[]
for i in hiResTest:
  lowResTest.append(cv2.resize(i,dsize=(128,128)))
```

```python
lowResTest = np.array(lowResTest)
categoriesTest = np.array(catsTest)
xTest = lowResTest/255
yTest = k.utils.to_categorical(categoriesTest)
```

```python
network = models.Sequential()
network.add(layers.Conv2D(64,(5,5), activation="relu", padding='same', input_shape=(128,128,3)))
network.add(layers.MaxPooling2D((2,2)))
network.add(layers.Conv2D(128,(4,4), activation="relu", padding='same'))
network.add(layers.MaxPooling2D((2,2)))
network.add(layers.Conv2D(256,(4,4), activation="relu", padding='same'))
network.add(layers.MaxPooling2D((2,2)))
network.add(layers.Flatten())
network.add(layers.Dense(5, activation = "softmax"))
```

```python
network.compile(optimizer='adam',
                loss='categorical_crossentropy',
                metrics=['accuracy'])
```

```python
network.summary()
```

```python
network.fit(xtrain, ytrain, batch_size = 50, epochs = 15)
```

```python
network.evaluate(xTest,yTest)
```
